# Data Interpolation
Interpolation is a mathematical technique to estimate the values of unknown data points that fall in between existing, known data points. In Python, there are various livraies available that can be uses for data inperpolation, such as NumPy, SciPy, and Pandas[link text](https://). Here is an example of how to perform data intrpolation using th Numpy libarary.

# Linear Interpolation

```python
import numpy as np
x=np.array([1,2,3,4,5])
y=np.array([2,4,6,8,10])
```

```python
import matplotlib.pyplot as plt
plt.scatter(x,y)
```

```python
## interpolate he data using linear interpolation
x_new = np.linspace(1,5,10) ## create new x values
y_interp = np.interp(x_new, x,y)
```

```python
plt.scatter(x_new, y_interp)
```

# Cubic Interpolation

```python
#Cubic Interpolation With Scipy
```

```python
x=np.array([1,2,3,4,5])
y=np.array([1,8,27,64,125])
```

```python
from scipy.interpolate import interp1d
```

```python
f= interp1d(x,y, kind ='cubic')
```

```python
x_new = np.linspace(1,5,10)
y_interp=f(x_new)
print(y_interp)
```

```python
plt.scatter(x_new, y_interp)
```

# Polynomial Interpolation

```python
 x=np.array([1,2,3,4,5])
y=np.array([1,2,9,15,25])
```

```python
# interploate the data using polynomial interpolation
p= np.polyfit(x,y,2)# fit a 2nd degree polynomial to the data
```

```python
x_new = np.linspace(1,5,10)#create new x values
y_interp = np.polyval(p, x_new) #intrpolate y values
```

```python
plt.scatter(x,y)
```

```python
plt.scatter(x_new, y_interp)
```
