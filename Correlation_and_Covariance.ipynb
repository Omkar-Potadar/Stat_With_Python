## Correlation and Covariance

```python
import seaborn as sns
```

```python
df=sns.load_dataset('healthexp')
df
```

```python
##covariance
df.cov()
```

```python
##Person correlation coefficient
df.corr(method='pearson')
```

```python
#spearman rank correlation
df.corr(method='spearman')
```
